#Ejercicio 1
Crear un módulo que grafique una portadora sinusoidal y otra que la module en amplitud.

```python
import numpy as np
import matplotlib.pyplot as plt

# Ejercicio 1
# Crear un módulo que grafique una portadora sinusoidal y otra que la module en amplitud.

def plot_modulated_signal(carrier_freq, modulating_freq, sampling_rate, duration, modulation_index):

    t = np.linspace(0, duration, int(sampling_rate * duration), endpoint=False)

    # Señal portadora
    carrier_signal = np.cos(2 * np.pi * carrier_freq * t)

    # Señal moduladora (para modulación de amplitud, la moduladora suele ser de menor frecuencia)
    modulating_signal = 0.5 * np.cos(2 * np.pi * modulating_freq * t) + 0.5 # Asegurarse de que la moduladora sea siempre positiva

    # Señal modulada en amplitud
    modulated_signal = carrier_signal * (1 + modulation_index * modulating_signal)


    plt.figure(figsize=(10, 6))

    plt.subplot(3, 1, 1)
    plt.plot(t, carrier_signal)
    plt.title('Señal Portadora')
    plt.xlabel('Tiempo (s)')
    plt.ylabel('Amplitud')
    plt.grid(True)

    plt.subplot(3, 1, 2)
    plt.plot(t, modulating_signal)
    plt.title('Señal Moduladora (para AM)')
    plt.xlabel('Tiempo (s)')
    plt.ylabel('Amplitud')
    plt.grid(True)

    plt.subplot(3, 1, 3)
    plt.plot(t, modulated_signal)
    plt.title('Señal Modulada en Amplitud')
    plt.xlabel('Tiempo (s)')
    plt.ylabel('Amplitud')
    plt.grid(True)

    plt.tight_layout()
    plt.show()

# Ejemplo de uso del Ejercicio 1
plot_modulated_signal(carrier_freq=100, modulating_freq=10, sampling_rate=1000, duration=1, modulation_index=0.8)
```
