# Ejercicio 2
Dada y = np.cos(n), mostrar los primeros 50 valores, reemplazando con 0 los negativos.

```python

import numpy as np
import matplotlib.pyplot as plt

n = np.arange(50)
y = np.cos(n*np.pi/10)
y_no_neg = y.clip(min=0)   # recorta por debajo de 0

print(y_no_neg)

plt.figure(figsize=(9, 4.8))
plt.plot(n, y_no_neg, marker='o', lw=1.8, label='cos(n) con negativos=0')
plt.axhline(0, color='gray', lw=0.8)
plt.title('Primeros 50 valores de cos(n) con negativos reemplazados por 0')
plt.xlabel('n (enteros)')
plt.ylabel('valor')
plt.grid(True, linestyle='--', alpha=0.35)
plt.legend()
plt.tight_layout()
plt.show()
```
