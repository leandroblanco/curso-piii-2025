# Análisis de Señal Senoidal con Muestreo y Cuantificación

Este notebook genera una señal senoidal de 1 kHz con amplitud de -5 a 5 V, la muestrea a 50 kHz, y muestra:
- La señal continua durante 10 ms
- Las primeras 50 muestras sin cuantificar
- Las mismas 50 muestras cuantificadas con un ADC de 12 bits

```python
import numpy as np
import matplotlib.pyplot as plt

frecuencia = 1000  # Hz
amplitud = 5
fs = 50000

duracion = 0.01
t = np.linspace(0, duracion, int(fs * duracion), endpoint=False)
senal_continua = amplitud * np.sin(2 * np.pi * frecuencia * t)

muestras = senal_continua[:50]
t_muestras = t[:50]

niveles_adc = 2**3
rango_adc = 2 * amplitud
resolucion = rango_adc / niveles_adc
senal_cuantificada = np.round((muestras + amplitud) / resolucion) * resolucion - amplitud

plt.figure(figsize=(12, 8))
plt.subplot(3, 1, 1)
plt.plot(t, senal_continua)
plt.title('Señal Senoidal Continua (10 ms)')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud [V]')

plt.subplot(3, 1, 2)
plt.stem(t_muestras, muestras)
plt.title('Primeras 50 Muestras sin Cuantificar')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud [V]')

plt.subplot(3, 1, 3)
plt.stem(t_muestras, senal_cuantificada)
plt.title('Primeras 50 Muestras Cuantificadas (ADC 12 bits)')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud [V]')

plt.tight_layout()
plt.show()
```
